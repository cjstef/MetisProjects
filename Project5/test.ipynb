```python
def tweets_to_df(list_of_tweets):
    emoticons_str = r"""
        (?:
            [:=;] # Eyes
            [oO\-]? # Nose (optional)
            [D\)\]\(\]/\\OpP] # Mouth
        )"""

    regex_str = [
        emoticons_str,
        r'<[^>]+>', # HTML tags
        r'(?:@[\w_]+)', # @-mentions
        r"(?:\#+[\w_]+[\w\'_\-]*[\w_]+)", # hash-tags
        r'http[s]?://(?:[a-z]|[0-9]|[$-_@.&amp;+]|[!*\(\),]|(?:%[0-9a-f][0-9a-f]))+', # URLs

        r'(?:(?:\d+,?)+(?:\.?\d+)?)', # numbers
        r"(?:[a-z][a-z'\-_]+[a-z])", # words with - and '
        r'(?:[\w_]+)', # other words
        r'(?:\S)' # anything else
    ]

    tokens_re = re.compile(r'('+'|'.join(regex_str)+')', re.VERBOSE | re.IGNORECASE)
    emoticon_re = re.compile(r'^'+emoticons_str+'$', re.VERBOSE | re.IGNORECASE)

    def tokenize(s):
        return tokens_re.findall(s)

    def preprocess(s, lowercase=False):
        tokens = tokenize(s)
        if lowercase:
            tokens = [token if emoticon_re.search(token) else token.lower() for token in tokens]
        return tokens
    
    unicodeChars = [u"\u2026",u'\u2013',u'\u2019', u'\ud83d', u'\ude02', u'\u2605',u'\ud83c','u\ude06']
    punctuation = list(string.punctuation)
    stop = stopwords.words('english') + punctuation + ['rt', 'via', 'bitcoin', 'btc'] + unicodeChars
        
    data = []
    temp = []
    hashes = []

    for item in list_of_tweets:
        cleaned_text_no_stops = []
        cleaned_text = preprocess(item['text'].lower())
        for word in cleaned_text:
            if word not in stop and not word.startswith(('#', '@', '&amp', 'http' )):
                cleaned_text_no_stops.append(word)
        pol = TextBlob(item['text']).sentiment.polarity
        subj = TextBlob(item['text']).sentiment.subjectivity
        
        temp.append(item['created_at'])
        temp.append(item['text'])
        if len(item['entities']['hashtags']) > 0:
            for i in item['entities']['hashtags']:
                hashes.append(i['text'])
        else:
            hashes = ''
        temp.append(hashes)
        temp.append((item['user']['name'], item['user']['screen_name']))
        temp.append(item['user']['followers_count'])
        temp.append(item['favorite_count'])
        temp.append(item['retweet_count'])
        temp.append(pol)
        temp.append(subj)
        temp.append(cleaned_text_no_stops)
        data.append(temp)
        temp = []
        hashes = []
    
    df = pd.DataFrame(data, columns=['timestamp','text','hashtags','user','user_followers','favorites','retweets','polarity','subjectivity','tokenized_text'])
    df['timestamp'] = pd.to_datetime(df.timestamp)
    return df
    
    
```

```python
import plotly.plotly as py
import plotly.figure_factory as ff

def info_dashboard(df):


    tweet_info_table = [['Significant Tweets', 'Tweet Text'],
                   ['Highest Followers User', df.text.iloc[df.user_followers.idxmax()].encode('utf-8')],
                   ['Most Popular', df.text.iloc[df.favorites.idxmax()].encode('utf-8')],
                   ['Most Polarizing (Positive)', df.text.iloc[df.polarity.idxmax()].encode('utf-8')],
                    ['Most Polarizing (Negative)', df.text.iloc[df.polarity.idxmin()].encode('utf-8')]]

    table = ff.create_table(tweet_info_table)

    def adjust_column_widths(ff_table, n_cols=2, new_xs=[.02]):
        for i,entry in enumerate(ff_table['layout']['annotations']):
            if i % n_cols!=0:
                entry['x'] = new_xs[(i % n_cols)-1]
    adjust_column_widths(table)

    return py.iplot(table, filename='simple_table')

def info_dashboard2(df):
    total_info = [['Scraped Data Overview', ''],
                  ['Total Tweets', len(df)], 
                 ['Timeframe', df.timestamp.max()-df.timestamp.min()]]


    table = ff.create_table(total_info)

 
    return py.iplot(table, filename='simple_table')
```

```python
import seaborn as sns
import pandas as pd
sns.set(color_codes=True)
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
from nltk import bigrams
from collections import Counter
import numpy as np
from sklearn import decomposition
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem import WordNetLemmatizer

def dashboard(df, timeframe='1min', num_topics=3, word_per_topic=5, top_topics_count=5):
        
    negative = positive = neutral = 0
    for item in df.polarity:
        if item == 0:
            pass
        elif item < -.33:
            negative+=1
        elif item > .33:
            positive+=1
        else:
            neutral+=1
    sentiments = [negative, neutral, positive]
    names = ['negative', 'neutral', 'positive']
    
    idx = pd.DatetimeIndex(df.timestamp)
    polarity_scores = pd.Series(list(df.polarity), index=idx)
    pol_per_timeframe = polarity_scores.resample(timeframe,how='mean').fillna(0)

    
    non_neutral_sentiments = []
    for i in range(len(df)):
        if 0 <= df.polarity.iloc[i] <.1:
            pass
        else:
            non_neutral_sentiments.append((df.polarity.iloc[i], df.subjectivity.iloc[i]))

    
    word_soup = []
    filter_words = ['amp']
    wordnet_lemmatizer = WordNetLemmatizer()
    for row in df.tokenized_text:
        for word in row:
            if word not in filter_words:
                word_l = wordnet_lemmatizer.lemmatize(word)
                word_soup.append(word_l)
            
        
    vectorizer = CountVectorizer()
    doc_term_matrix = vectorizer.fit_transform(word_soup).toarray()
    
    clf = decomposition.NMF(n_components=num_topics, random_state=1)
    
    doctopic = clf.fit_transform(doc_term_matrix)
    
    vocab = np.array(vectorizer.get_feature_names())

    topic_words = []
    
    for topic in clf.components_:
        word_idx = np.argsort(topic)[::-1][0:word_per_topic]
        topic_words.append([vocab[i] for i in word_idx])
        
    i = 1
    print 'NMF Topic Discovery:'
    for t in topic_words:
        print 'topic ' + str(i) + ': ' + ', '.join(t)
        i+=1
    
    count_all = Counter()
    count_all.update(word_soup)
    plotdata = count_all.most_common(top_topics_count+1)
    plotdata.pop(0)
    
    topics =[]
    for item in plotdata:
        if type(item[0]) == tuple:
            topics.append(item[0][0]+' '+item[0][1])
        else:
            topics.append(item[0])
    mentions = zip(*plotdata)[1]
    x_pos = np.arange(len(topics)) 
    
    df_sentiments = pd.DataFrame(non_neutral_sentiments, columns = ['polarity','subjectivity'])

    f, (ax1, ax2) = plt.subplots(2,figsize=(6, 6))
    sns.distplot(df_sentiments.polarity, ax=ax2)
    sns.barplot(x=names, y=sentiments,ax=ax1)
    #sns.tsplot(pol_per_timeframe,ax=ax3)
    #sns.barplot(x=x_pos, y=mentions, ax=ax4)
    sns.jointplot(x='polarity', y='subjectivity', data=df_sentiments, kind="kde", space=0, color="k")
    sns.plt.title('Net Sentiment Density', fontweight='bold',fontsize=14)
    ax1.set_title('Sentiment Category vs Tweet Count', fontweight='bold',fontsize=14)
    ax2.set_title('Sentiment Score Distribution', fontweight='bold',fontsize=14)
    #ax3.set_title('Aggregate Sentiment Per ' + str(timeframe), fontweight='bold',fontsize=14)
    #ax4.set_title('Common topics by word count', fontweight='bold',fontsize=14)
    #ax4.set(xlabel='Topic',ylabel='# of mentions')
    #ax4.xaxis.set_ticklabels(topics,rotation=45, fontweight='bold',fontsize=12)
    f.subplots_adjust(hspace=.3)
    
    a4_dims = (7, 5)
    fig, ax = plt.subplots(figsize=a4_dims)
    ax.set_title('Topics by Word Count')
    ax.set(xlabel='Topic',ylabel='Num. of Mentions')
    sns.barplot(ax=ax, x=x_pos, y=mentions)
    ax.xaxis.set_ticklabels(topics, rotation=45)
    return df




```
